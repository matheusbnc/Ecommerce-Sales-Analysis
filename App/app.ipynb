```python
from menus.menus import *
from results.produtos import *
from results.clientes import *
from results.paises import *
from results.sazonalidade import *
from preprocessing.formatting import title_col, change_types, pre_processing, drop_nan\
                                        , drop_uk, add_time_columns
from time import sleep
import pandas as pd

 # Função principal para executar o programa
def main():
    
    retail = pd.read_csv('Data/Online Retail.csv', delimiter=';', encoding='latin1', decimal=',')
    retail = pre_processing(retail)
    change_types(retail)

    while True:
        main_menu()
        sleep(0.5)
        opcao = input('\nEscolha uma opção: ')
        sleep(0.5)
        
        if opcao == '1':
            while True:
                
                sleep(0.5)
                produto_submenu()
                opcao_produto = input('\nEscolha uma opção: ')
                
                if opcao_produto == '1':
                    more_units_sold(retail)
                    
                elif opcao_produto == '2':
                    more_profitable(retail)
                    
                elif opcao_produto == '3':
                    frequent_sales(retail)
                    
                elif opcao_produto == '4':
                    more_units_canceled(retail)
                    
                elif opcao_produto == '5':
                    expensive_cancellations(retail)
                
                elif opcao_produto == '6':
                    frequent_cancellations(retail)

                elif opcao_produto == '0':
                    break
                    
                else:
                    print('\nOpção inválida. Por favor, escolha uma opção válida.\n')

        elif opcao == '2':

            retail_cli = drop_nan(retail)
            
            while True:
                
                sleep(0.5)
                cliente_submenu()
                opcao_cliente = input('\nEscolha uma opção: ')
                
                if opcao_cliente == '1':
                    more_units_sold_clients(retail_cli)
                    
                elif opcao_cliente == '2':
                    more_profitable_clients(retail_cli)
                    
                elif opcao_cliente == '3':
                    frequent_clients(retail_cli)
                
                elif opcao_cliente == '4':
                    more_units_canceled_clients(retail_cli)
                    
                elif opcao_cliente == '5':
                    expensive_cancellations_clients(retail_cli)
                    
                elif opcao_cliente == '6':
                    frequent_cancellations_clients(retail_cli)
                    
                elif opcao_cliente == '0':
                    break
                    
                else:
                    print('\nOpção inválida. Por favor, escolha uma opção válida.\n')
                    
        elif opcao == '3':
            
            retail_ctr = drop_uk(retail)
            
            while True:
                
                sleep(0.5)
                pais_submenu()
                opcao_pais = input('\nEscolha uma opção: ')
                
                if opcao_pais == '1':
                    more_units_sold_country(retail_ctr)
                    
                elif opcao_pais == '2':
                    more_profitable_country(retail_ctr)
                    
                elif opcao_pais == '3':
                    frequent_sales_country(retail_ctr)
                    
                elif opcao_pais == '4':
                    more_units_canceled_country(retail_ctr)
                    
                elif opcao_pais == '5':
                    expensive_cancellations_country(retail_ctr)
                    
                elif opcao_pais == '6':
                    frequent_cancellations_country(retail_ctr)
                    
                elif opcao_pais == '0':
                    break
                    
                else:
                    print('\nOpção inválida. Por favor, escolha uma opção válida.\n')
                    
        elif opcao == '4':
            
            retail_saz =  add_time_columns(retail)
            retail_saz = retail_saz[retail_saz['Year'] == 2011] # Filtra por 2011, para não repetir mês
            while True:
                
                sleep(0.5)
                sazonalidade_submenu()
                opcao_sazonalidade = input('\nEscolha uma opção: ')
                
                if opcao_sazonalidade == '1':
                    daily_variation_quantity(retail_saz)
                    
                elif opcao_sazonalidade == '2':
                    daily_variation_price(retail_saz)
                    
                elif opcao_sazonalidade == '3':
                    daily_variation_frequency(retail_saz)
                    
                elif opcao_sazonalidade == '4':
                    monthly_variation_quantity(retail_saz)
                    
                elif opcao_sazonalidade == '5':
                    monthly_variation_price(retail_saz)
                    
                elif opcao_sazonalidade == '6':
                    monthly_variation_frequency(retail_saz)
                    
                elif opcao_sazonalidade == '7':
                    quarterly_variation_quantity(retail_saz)
                    
                elif opcao_sazonalidade == '8':
                    quarterly_variation_price(retail_saz)
                    
                elif opcao_sazonalidade == '9':
                    quarterly_variation_frequency(retail_saz)
                    
                elif opcao_sazonalidade == '0':
                    break
                    
                else:
                    print('\nOpção inválida. Por favor, escolha uma opção válida.\n')
                    
        elif opcao == '0':
            print('\nSaindo do programa...')
            break

        else:
            print('\nOpção inválida. Por favor, escolha uma opção válida.\n')

if __name__ == '__main__':
    main()
```
